```python
student_grades_df = df.copy()
student_grades_df
```

```python
#multiple columns dropping at one time
# student_grades_df = student_grades_df.drop(['G1','G2','G3'], axis = 1)
# student_grades_df
```

```python
#Q1 - 1.5*IQR
#Q3 + 1.5*IQR
```

```python
def detect_outliers(columns):
    outlier_indices = []
    for column in columns:
        Q1 = student_grades_df[column].quantile(0.25)
        Q3 = student_grades_df[column].quantile(0.75)
        IQR = Q3 - Q1

        #mask is a boolean condition
        mask = (student_grades_df[column] >= (Q1 - 1.5*IQR)) & (student_grades_df[column] <= (Q3 + 1.5*IQR))
        mask = mask.to_numpy() #bool to numpy array 

        #find the indices of values that do not satisfy the mask condition that will be our outlier
        false_indices = np.argwhere(~mask)
        outlier_indices.append(false_indices)

    return np.unique(np.concatenate(outlier_indices).ravel())
        
#np.concatenate(outlier_indices)>> combine all the outlier index arrays into a single array
#so each column will have one array
#.ravel>> flattens the multi-dimensional array into 1 d array        
```

```python
num_cols = ["age", "absences"]
detect_outliers(num_cols)
```

```python
cat_features = X.select_dtypes(include = "object").columns
num_features = X.select_dtypes(exclude = "object").columns
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

from sklearn.compose import ColumnTransformer

oh_transformer = OneHotEncoder(drop = "first")  #we know that in OHE 3rd feature is unnecessary to drop that -> drop = "first"
numeric_transformer = StandardScaler()

preprocessor = ColumnTransformer(
    [
    ("OneHotEncoder",oh_transformer,cat_features),
    ("StandardScaler",numeric_transformer,num_features)
])
```

```python
preprocessor
```

```python
preprocessor.fit_transform(X_train)
```
